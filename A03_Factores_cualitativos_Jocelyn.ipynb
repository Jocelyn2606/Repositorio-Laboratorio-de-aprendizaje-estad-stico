```python
import pandas as pd
from google.colab import files

df = pd.read_excel("Elfeo (4).xlsx")
df.head()
```

Ahora la tabla la transformamos en 0 y 1

grande= 1
Chico = 0

Sin = 0
con = 1

```python
df['Tamaño']= df['Tamaño'].replace({"Grande":1 , "Chica":0})
df['Hielo']= df['Hielo'].replace({"Con":1 , "Sin":0})
```

```python
df["Kahlúa"] = (df["Sabor"] == "Kahlúa").astype(int)
df["Horchata"] = (df["Sabor"] == "Horchata").astype(int)
```

```python
df
```

# Parte 1

**Regresión Lineal**

```python
import statsmodels.api as sm
```

```python
xf = df[["Tamaño","Hielo","Kahlúa","Horchata"]]
xf1 = sm.add_constant(xf)
```

```python
y = df["Score"].values.reshape(-1,1)
```

```python
xff = sm.add_constant(xf1)
model = sm.OLS(y,xff)
results = model.fit()
results.summary()
```
