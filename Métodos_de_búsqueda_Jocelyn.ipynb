```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.preprocessing import StandardScaler
```

```python
x = np.array([10, 2, 3, 7, 12]).reshape(-1, 1)
y = np.array([2847, -4017, -1255, 1773, 563])
```

```python
kernel = 1.0 * RBF(length_scale=1.0)
```

```python
gp = GaussianProcessRegressor(n_restarts_optimizer=10)
gp.fit(x, y)
```

```python
xsim = np.linspace(2,15,10001).reshape(-1,1)
```

```python
y_predict, sigma= gp.predict(xsim,return_std=True)
y_predict
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y, color="red", label="Datos")
plt.plot(xsim, y_predict, color="blue", label="Predicción")

plt.fill_between(
    xsim.ravel(),
    y_predict - 2*sigma,
    y_predict + 2*sigma,
    color="blue",
    alpha=0.2,
    label="Incertidumbre"
)
plt.ylim([0,3000])
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gaussian Process Regression")
plt.show()
```

```python
imax=np.argmax(y_predict+2*sigma)
imax
#en este caso si solo ponemos y_predict da el mismo valor
```

```python
xsim[imax]
```
