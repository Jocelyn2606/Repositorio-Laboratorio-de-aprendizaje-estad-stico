**1.1 Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Considera todos los demás factores como numéricos/ordinales.**

```python
#Eliminamos columna "model"
df = df.drop("model", axis=1)
df.head()
```

*   Calcula el R2 e interpreta los signos de los betas.


```python
x = df.drop(columns=["mpg"]).values
```

```python
y = df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
intercepto = lr.intercept_
coef = lr.coef_
print("Intercepto:", intercepto)
print("Coeficientes:", coef)
```

```python
r2 = lr.score(x,y)
print("R^2:", r2)
```

```python
df.columns
```

*Interpretación:*

Nuestro R2 tiene un valor de 86.9%, lo cual nos indica que el modelo explica 86% de la variabilidad.

Respecto a los cieficientes, tenemos alguno negativos como cyl, hp (peso), wt (caballos de fuerza) y carb.El tener un negativo nos dice que cada vez que esta variable aumente las millas por galon van a disminuir. Por el otro lado, al tener un valor postivo, esto nos dice que hay una mayor relación. Por ejemplo, am (tipo de transmisión) tiene un variable porsitiva lo cual nos demuestra que si esta variable aumenta , también lo harán las millas por galon.

*   Realiza un train-test-split donde se use el 40% de los datos para entrenar. Calcula el R2 de entrenamiento y de prueba.



```python
len(df)*.4
```

```python
#split train
X_train = x[:12]
y_train = y[:12]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
r2_train = lr.score(X_train,y_train)
print("R2 train:", r2_train)
```

```python
#split test
X_test = x[12:]
y_test = y[12:]
```

```python
r2_test = lr.score(X_test,y_test)
print("R2 test:", r2_test)
```

El modelo en r2_train obtuvo un 99% y el r2_test obtuvo un -49% lo cual nos dice que no se esta generalizando porque hay mucha diferencia entre estos. Hay un sobreajuste.

*   Añade regularización L2 con un hiperparámetro lambda decidido por ti. Cambia este valor y compara con varios distintos los R2 de entrenamiento y de prueba.

```python
#Ridge

from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha=0.1)
ridge.fit(X_train, y_train)

r2_train = ridge.score(X_train, y_train)
r2_test = ridge.score(X_test, y_test)
print("R2 train:", r2_train)
print("R2 test:", r2_test)
```

```python
ridge = Ridge(alpha=1)
ridge.fit(X_train, y_train)

r2_train = ridge.score(X_train, y_train)
r2_test = ridge.score(X_test, y_test)
print("R2 train:", r2_train)
print("R2 test:", r2_test)
```

```python
ridge = Ridge(alpha=10)
ridge.fit(X_train, y_train)

r2_train = ridge.score(X_train, y_train)
r2_test = ridge.score(X_test, y_test)
print("R2 train:", r2_train)
print("R2 test:", r2_test)
```

*Interpretación:*

como se puede observar, al agregarle una regularización L2 el modelo se va ajustando y deja de considerarse un modelo sobre ajustado. R2 teste dejó de ser negativo y re train bajo un poco. Hay una mejora en la generalización del modelo.

**2.1 Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Crea columnas dummies para los factores 'cyl', 'gear' y 'carb'.**

```python
df2 = df.copy()
```

```python
# "cyl"
df2["cyl_6"] = (df2["cyl"] == 6).astype(int)
df2["cyl_8"] = (df2["cyl"] == 8).astype(int)
```

```python
df2 = df2.drop("cyl", axis=1)
```

```python
df2["gear"].unique()
```

```python
# "gear"
df2["gear_4"] = (df2["gear"] == 4).astype(int)
df2["gear_5"] = (df2["gear"] == 5).astype(int)
df2 = df2.drop("gear", axis=1)
```

```python
df2["carb"].unique()
```

```python
df2["carb_2"] = (df2["carb"] == 2).astype(int)
df2["carb_3"] = (df2["carb"] == 3).astype(int)
df2["carb_4"] = (df2["carb"] == 4).astype(int)
df2["carb_6"] = (df2["carb"] == 6).astype(int)
df2["carb_8"] = (df2["carb"] == 8).astype(int)

df2 = df2.drop("carb", axis=1)
```

```python
df2.head()
```



*   Calcula el R2 e interpreta los signos de los betas.


```python
X_d = df2.drop("mpg", axis=1).values
y_d = df2["mpg"].values
```

```python
lr_d = LinearRegression()
lr_d.fit(X_d, y_d)
```

```python
intercepto_d = lr_d.intercept_
coef_d = lr_d.coef_
print("Intercepto:", intercepto_d)
print("Coeficientes:", coef_d)
```

```python
r2_d = lr_d.score(X_d, y_d)
print("R2:", r2_d)
```

*Interpretación*

Gracias a que convertimos las variables categóricas en variables dummies, el modelo mejora su ajuste, en el primer r2 su valor era de 86% y en este es de 89%.
